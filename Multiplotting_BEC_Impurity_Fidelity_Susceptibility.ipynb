```python
####### Code by Matt Richards ###################################
####
#### This code takes data from a time dynamics simulation 
#### of a BEC-impurity double well model and creates a multiplot
################################################################

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
import matplotlib

############################################
### Modify font and text for readability ###
############################################

font = {'family' : 'serif',
        'style'  : 'italic',
        'weight' : 'normal',
        'size'   : 16}

matplotlib.rc('font', **font)
plt.rcParams['mathtext.fontset'] = 'dejavuserif'

### Take data from directory ###
dir = "./"
file = "eigs_comparison.txt"
fn=np.loadtxt(dir+file, dtype=np.double)
nowcol = 'r'
nowpt = 'o'

### Initialize Multiplot ###
fig, ax1 = plt.subplots()

### I am specifying the values I wish to look at from the data ###
k=[0 for i in range(10)]

k[1]=10
k[2]=50
k[3]=100
k[4]=200
k[5]=300
         
### From the data, the values in the loop are chosen so that they are non sensical ###
### Some of the data breaks down due to degeneracy in the eigenvectors ###
for i in range(1,6):
    ax1.plot(fn[i*101+300*(i-1):i*300+(i-1)*85,9], fn[i*101+300*(i-1):i*300+(i-1)*85,8], ls='--',label='N=%s'%(k[i]))
    

plt.axvline(x=2,color='k',label=r'$W=W_c$')
nowcol = 'b'
nowpt = 'o'

### Plotting options for the data ###
#plt.title("Fidelity Susceptibility for the BEC-impurity model")
plt.xlabel("$W(J)$")
plt.ylabel(r"$\chi_F (W)$")#r"$F(W)=|<E0(W)|E0(W)+\delta W>|^2$")#
plt.tight_layout()
plt.legend(loc=(0.7,0.3),  borderaxespad=0, prop={'size':12})
plt.yticks(np.arange(0,400, step=100))
plt.ticklabel_format(useOffset=False)

### Plotting options for the multiplot ###
left, bottom, width, height = [0.3, 0.35, 0.2, 0.4] #This needs to be explicitly written after the first plot is generated 
#to indicate that the plot commands are given for the main plot.
ax2 = fig.add_axes([left, bottom, width, height])

#plt.xticks(np.arange(0, 4.0, step=0.5))
ax2.plot(fn[101:400,9],fn[101:400,8], ls='--')
ax2.ticklabel_format(useOffset=False)
ax2.set_xticks(np.arange(1.0, 4.1, step=1.0))

plt.show()
#plt.savefig("./"+'chiW_defence.jpg',dpi=300)
```
